```python
## Dropping NaN countries

tax_haven = tax_haven.dropna(subset=["country_name"])
fin_secrecy = fin_secrecy.dropna(subset=["country_name"])
```

```python
## According to SSPI Indicator Info, we are looking at "score of the world’s most important tax havens".
## Hence, I will drop other features like rankings and global scale weight and focus on tax haven score.

tax_haven_score = tax_haven[["iso3","cthi_2019_score", "cthi_2021_score"]]
tax_haven_score
```

```python
## Similarly clean fin_secrecy and only include data for 2019 and 2021.
cols_to_keep = ["iso3"] + [col for col in fin_secrecy.columns.to_list() if "score" in col]
fin_secrecy_score = fin_secrecy[cols_to_keep]
fin_secrecy_score
```

```python
tax_eva_df = pd.merge(tax_haven_score, fin_secrecy_score, on="iso3", how="inner")
tax_eva_df
```

```python
## Drop countries with no data
tax_eva_df = tax_eva_df.dropna(how="all")
tax_eva_df
```

```python
tax_eva_df = tax_eva_df.iloc[:,[0,3,4,5,6,1,7,2,8]]
tax_eva_df
```

```python
## Formatting
tax_eva_df.insert(0, "IndicatorCode", "TAXEVA")
tax_eva_df = tax_eva_df.rename(columns={"iso3" : "CountryCode"})
tax_eva_df
```

```python
import re

def extract_year_and_type(col):
    matches = re.match(r'(fsi|cthi)_(\d+)_score', col)
    if matches:
        type_ = matches.group(1).upper()  # 'FSI' or 'CTHI'
        year = matches.group(2)  # Year as string
        return type_, year
    return None, None

new_cols = {}
types = {}

for col in tax_eva_df.columns:
    score_type, year = extract_year_and_type(col)
    if year:
        new_cols[col] = year
        if score_type == "FSI":
            types[year] = "FINSEC"
        elif score_type == "CTHI":
            types[year] = "TAXHAV"

tax_eva_df.rename(columns=new_cols, inplace=True)
tax_eva_df
```

```python
tax_eva_melted = pd.melt(tax_eva_df, id_vars=["IndicatorCode", "CountryCode"], var_name="Year", value_name="Value")
tax_eva_melted["Unit"] = "Score"
tax_eva_melted
```

```python
descriptions = {"FINSEC": "Score based on 20 measures of financial secrecy, grouped around ownership registration, legal entity transparency, integrity of tax and financial regulation, and international standards and cooperation.",
                "TAXHAV": "Score of the world’s most important tax havens, rated by how aggressively and extensively each jurisdiction helps multinationals evade taxes."}

def intermediate(row):
  intermediates = {}
  intermediates["CountryCode"] = row["CountryCode"]
  intermediates["Description"] = descriptions[types[str(row["Year"])]]
  intermediates["IndicatorCode"] = types[row["Year"]]
  intermediates["Value"] = row["Value"]
  intermediates["Unit"] = "Score"
  intermediates["Year"] = row["Year"]
  return intermediates

tax_eva_melted["Intermediate"] = tax_eva_melted.apply(intermediate, axis=1)
tax_eva= tax_eva_melted.iloc[:,[0,1,5,2,3,4]]
tax_eva
```

```python
## Export the cleaned bulk data in dictionary format

tax_eva.to_json('tax_eva.json', orient="records", indent=4)
```
